# Resizing 2D Arrays

_I will demonstrate how to resize arrays. Maybe useful for spatially resolved studies of galaxies._

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import signal
```

I will create a random 2D array. We define the original array size and the final array size.

```python
plt.rcParams.update({'figure.dpi': 200})

# Define the parameters
originalsize = 100
resamplesize = 10

a_orig = np.zeros((originalsize,originalsize))
for (y,x), value in np.ndenumerate(a_orig):
    a_orig[y,x] = (y-x)

plt.imshow(a_orig)
plt.title("%s samples each side"%originalsize)
```

# Resampling the array

There are many techniques. I show the easiest method.

## Fourier method
This technique will calculate the Fourier transform, truncate the Fourier components and inverse Fourier transform.

```python
a_resample = a_orig.copy()
a_resample = signal.resample(a_resample, num=resamplesize, axis=0) # resizing in the direction of each axis
a_resample = signal.resample(a_resample, num=resamplesize, axis=1) # resizing in the direction of each axis

plt.imshow(a_resample)
plt.title("Resampled to %s samples each side"%resamplesize)
```
